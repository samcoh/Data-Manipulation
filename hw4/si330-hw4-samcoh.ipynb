We have provided you with an SQLite database in `chinook.db`. This database contains the following tables:

- `artists`: Musical artists, each with an artist id and name.

```python
# CHANGE ME: insert code here
statement = '''
SELECT art.Name as Artist_Name, Count(alb.AlbumId) as Number_of_Albums 
FROM artists as art 
JOIN albums as alb
ON alb.ArtistId = art.ArtistId
GROUP BY alb.ArtistID
ORDER BY Count(alb.AlbumId) DESC
LIMIT 10
'''

cur.execute(statement)
[dict(row) for row in cur.fetchall()]
```

**2.10** How many "Alternative & Punk" songs were purchased in the United States?

```python

statement = '''
SELECT genres.Name as Genre_Name,COUNT(invoice_items.TrackId) as Number_of_Purchases
FROM tracks
JOIN genres
ON genres.GenreId = tracks.GenreId
JOIN invoice_items 
ON tracks.TrackId = invoice_items.TrackId
JOIN invoices 
ON invoice_items.InvoiceId = invoices.InvoiceId
WHERE genres.Name = "Alternative & Punk" AND invoices.BillingCountry = "USA"
GROUP BY genres.Name
'''
cur.execute(statement)
[dict(row) for row in cur.fetchall()]
```
