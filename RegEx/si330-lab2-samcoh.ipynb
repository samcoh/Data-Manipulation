### Part 1.4

In the above code chunk you created a dictionary whose keys are tuples. However, tuples are an ordered list. Hence, if there were emails between two person, 'A' and 'B', your dictionary might have ('A', 'B') and ('B', 'A') as separate keys. At this point, we don't really care about sender and receiver; we just want to know if two people have communicated with each other. Thus, what you need to do here is to sort your tuple so that the email ids are always in alphabetical order. You will modify your code from the previous chunk and implement a sorting function before you store a key value pair in your dictionary.

<font color="magenta">Repeat your regular expression from the previous chunk which gives you a tuple. <b>Use the `sorted` function to sort the tuple in alphabetical order first</b>, and then store this in a dictionary as you did previously. The sorted function outputs a list. Make sure you convert your list to a tuple using `tuple()`.</font>

````python
email_senders_receivers = {} # CHANGE ME: initialize this to an appropriate data structure
for email in email_data:
    from_to_pattern = re.compile(r'From:\s+([a-zA-Z0-9.-_]+)@.*\nTo:\s([a-zA-Z0-9.-_]+)@', re.DOTALL)
    match = from_to_pattern.search(email['message'],re.DOTALL)
    if match:
        sorted_tuple_key = sorted(match.groups(), key = lambda x:x)
        sorted_tuple_key = (sorted_tuple_key[0], sorted_tuple_key[1])
        if sorted_tuple_key in email_senders_receivers: 
            email_senders_receivers[sorted_tuple_key] += 1
        else: 
            email_senders_receivers[sorted_tuple_key] = 1 
            
````
