## Survey says...

Last year, students in several courses filled out a survey at the beginning of the course where they were asked questions such as Expected Grade, Expected Number of Hours, Enthusiasm, Worry and some questions about their proficiency in Python. In this part, we will use it to do some simple manipulation tasks using Pandas.

````python
df = pd.read_csv('deidentified-survey.csv')
````

Let's check what the data looks like.

````python
df.head(n = 2)
````

The column names look a bit lengthy, and it's easy to make a typo when you try to use them. We can rename them:

````python
# use '\' to break a long line into multiple lines make your code much more readable.
df.columns = ['firstName','lastName','expectedGrade','planedHours',\
              'enthusiasm','worry','q1','q2','q3','q4','q5', 'overallScore']
````

Let's check the data again to make sure the columns are renamed as expected.

**Q0.1.** Retrieve the last 3 rows in the datafram and output these columns: `firstName`, `lastName`, `expectedGrade`, `enthusiasm`, `worry`.

````python
#df[["firstName", "lastName","expectedGrade", "enthusiasm", "worry"]].tail(3)
df.tail(3)[["firstName", "lastName","expectedGrade", "enthusiasm", "worry"]]
````

The following line creats a new column `fullName` to store the full names of the students in the dataframe by combining two columns: ```firstName``` and ```lastName```. 

````python
df['fullName'] = df['firstName'] + " " + df['lastName']
````

In the survey, students filled a `5 - point` _Likert Scale_ with values from 1 - 5 to describe how worried they were about the course. 

We want to calculate the relaxation score such that a person whose worry score is 5 would have a relaxation score of 1. Correspondingly, for someone whose worry score is 2, their relaxation score would be 4.

**Q0.2.** Calculate the relaxation score of each student and store it in the original dataframe with column name `relaxation`.

````python
df["relaxation"] = 6 - df["worry"]
#df[["relaxation","worry"]]
````

**Q0.3.** Find out the students who are either too relaxed about the course (relaxation scores > 4) or too stressed out (relaxation scores < 2). Only output the full name and relaxation score of those students (use one line of code).

````python
df[(df["relaxation"]>4) | (df["relaxation"]<2)][["fullName","relaxation"]]
````

__Bonus__: Use one line of code to get the `fullName`, `expectedGrade`, `relaxation` score of the following students based on first name: September, Lan, Candyce, Garth. (Hint: you can use ``.isin()`` function)

````python
df[(df["firstName"].isin(["September", "Lan", "Candyce", "Garth"]))][["fullName","expectedGrade","relaxation"]]

````

**Q1.2.** Print out a list of how many people were killed by each natural disaster type (including all natural disasters) in 1992.

````python
#disasters.Entity.value_counts()
disasters[(disasters["Year"] == 1992)][["Deaths","Entity"]]
````

**Q1.3.** Print out a list of how many people were killed by each natural disaster type (including all natural disasters) in 1992, **sorted by the number of deaths (largest first)**

````python
disasters[(disasters["Year"] == 1992)][["Deaths","Entity"]].sort_values(["Deaths"],ascending = False)
````

**Q1.4.** List how many people were killed by **Wildfire** in 1992.

````python
disasters[(disasters["Year"] == 1992)&(disasters["Entity"]=="Wildfire")][["Deaths"]]
````

**Q1.5.** List how many people were killed by **Wildfire** in each year:

````python
disasters[(disasters["Entity"]=="Wildfire")][["Year","Deaths"]]
````

**Q1.6.** How many people in total were killed by wildfires in this dataset? (i.e., give the total number of deaths from wildfire in this dataset: this should be just one number).

*Hint:* use `.sum()`

````python
disasters[(disasters["Entity"]=="Wildfire")][["Deaths"]].sum()
````

**Q1.7.** How many people in total were killed by each disaster type in this dataset? This should output a Series with 11 values.

*Hint:* use `.groupby()`

````python
disasters.groupby(["Entity"])["Deaths"].sum()
#disasters[["Entity","Deaths"]].groupby('Entity').sum()
````
